```python
def trainContextAndEvaluation(file,time,train_feat, test_feat, train_label, threshold_non, threshold_abu, type_class,unlabeled_feat):
    print('train_context_feature')
    print(train_feat.shape[0])
    print(train_label.shape)
    certain_unlabeled_data = [];
    import sklearn.ensemble as ensemble
    import numpy
    from imblearn.ensemble import BalancedBaggingClassifier
    from sklearn.ensemble import RandomForestClassifier
    
    model = RandomForestClassifier(n_estimators=20, n_jobs=20)
    
    #except df[0] that is comment_id
    train_data = pd.DataFrame.sparse.from_spmatrix(train_feat)
    train_data = train_data[range(1,train_data.shape[1])]
    
    test_data = pd.DataFrame.sparse.from_spmatrix(test_feat)
    test_data = test_data[range(1,test_data.shape[1])]
    
    unlabeled_data = pd.DataFrame.sparse.from_spmatrix(unlabeled_feat)
    unlabeled_data = unlabeled_data[range(1,unlabeled_data.shape[1])]
    
    model.fit(train_data, train_label)
    predictions_test = model.predict_proba(test_data)
    predictions_train = model.predict_proba(train_data)
    predictions_unlabeled = model.predict_proba(unlabeled_data)
    
    for th in numpy.arange(0.1,1.1,0.1):
        predictions_class_Train = (predictions_train[:,1] >= th).astype('int')
        predictions_class_Test = (predictions_test[:,1] >= th).astype('int')

        report_performance(file,time,"context",predictions_class_Train,predictions_class_Test, type_class,th, predictions_train[:,1], predictions_test[:,1],train_label,train_feat.shape[0])

    
    for i, v in enumerate(predictions_unlabeled[:,1]):
        if(v >= threshold_abu):
                certain_unlabeled_data.append([1])

        elif(v <= threshold_non):
                certain_unlabeled_data.append([0])

        elif(v > threshold_non and v < threshold_abu):
                certain_unlabeled_data.append([0.5])

    return certain_unlabeled_data
            
    
```

```python
def trainContentAndEvaluation(file,time,train_feat, test_feat, train_label, threshold_non, threshold_abu, type_class,unlabeled_feat):
    print('train_content_feature')
    print(train_feat.shape[0])
    print(train_label.shape)
    certain_unlabeled_data = [];
    import sklearn.ensemble as ensemble
    import numpy
    from imblearn.ensemble import BalancedBaggingClassifier
    from sklearn.ensemble import RandomForestClassifier
    
    model = RandomForestClassifier(n_estimators=20, n_jobs=20)
    
    #except df[0] that is comment_id
    train_data = pd.DataFrame.sparse.from_spmatrix(train_feat)
    train_data = train_data[range(1,train_data.shape[1])]
    
    test_data = pd.DataFrame.sparse.from_spmatrix(test_feat)
    test_data = test_data[range(1,test_data.shape[1])]
    
    unlabeled_data = pd.DataFrame.sparse.from_spmatrix(unlabeled_feat)
    unlabeled_data = unlabeled_data[range(1,unlabeled_data.shape[1])]

    model.fit(train_data, train_label)
    predictions_test = model.predict_proba(test_data)
    predictions_train = model.predict_proba(train_data)
    predictions_unlabeled = model.predict_proba(unlabeled_data)
    
    for th in numpy.arange(0.1,1.1,0.1):
        predictions_class_Train = (predictions_train[:,1] >= th).astype('int')
        predictions_class_Test = (predictions_test[:,1] >= th).astype('int')

        report_performance(file,time,"content",predictions_class_Train,predictions_class_Test, type_class,th, predictions_train[:,1], predictions_test[:,1],train_label,train_feat.shape[0])

        
    for i, v in enumerate(predictions_unlabeled[:,1]):
        if(v >= threshold_abu):
                certain_unlabeled_data.append([1])
        elif(v <= threshold_non):
                certain_unlabeled_data.append([0])
        elif(v > threshold_non and v < threshold_abu):
                certain_unlabeled_data.append([0.5])
    return certain_unlabeled_data
            
    
```

```python
def predict_class(list_predict_prob,threshold):
    list_predict_class = [];
    for i in range (len(list_predict_prob)):
        if(list_predict_prob[i][1] > threshold):
            list_predict_class.append(1)
        else :
            list_predict_class.append(0)
    return list_predict_class
```

```python
def trainCombinedAndEvaluation(file,time,train_feat_context, train_feat_content, test_feat_context, test_feat_content, train_label_context, train_label_content, type_class, unlabeled_feat_context, unlabeled_feat_content):
    print('train_both_content_and_context')
    import sklearn.ensemble as ensemble
    import numpy
    from sklearn.ensemble import RandomForestClassifier
    for alpha in numpy.arange(0.0,1.1,0.1):
        model_context = RandomForestClassifier(n_estimators=20, n_jobs=20)
        model_content = RandomForestClassifier(n_estimators=20, n_jobs=20)

        train_context_data = pd.DataFrame.sparse.from_spmatrix(train_feat_context)
        train_context_data = train_context_data[range(1,train_context_data.shape[1])]
        train_content_data = pd.DataFrame.sparse.from_spmatrix(train_feat_content)
        train_content_data = train_content_data[range(1,train_content_data.shape[1])]

        test_context_data = pd.DataFrame.sparse.from_spmatrix(test_feat_context)
        test_context_data = test_context_data[range(1,test_context_data.shape[1])]
        test_content_data = pd.DataFrame.sparse.from_spmatrix(test_feat_content)
        test_content_data = test_content_data[range(1,test_content_data.shape[1])]

        unlabeled_context_data = pd.DataFrame.sparse.from_spmatrix(unlabeled_feat_context)
        unlabeled_context_data = unlabeled_context_data[range(1,unlabeled_context_data.shape[1])]
        unlabeled_content_data = pd.DataFrame.sparse.from_spmatrix(unlabeled_feat_content)
        unlabeled_content_data = unlabeled_content_data[range(1,unlabeled_content_data.shape[1])]

        model_context.fit(train_context_data, train_label_context)
        predictions_test_context = model_context.predict_proba(test_context_data)
#         predictions_train_context = model_context.predict_proba(train_context_data)
        predictions_unlabeled_context = model_context.predict_proba(unlabeled_context_data)

        model_content.fit(train_content_data, train_label_content)
        predictions_test_content = model_content.predict_proba(test_content_data)
#         predictions_train_content = model_content.predict_proba(train_content_data)
        predictions_unlabeled_content = model_content.predict_proba(unlabeled_content_data)

        for threshold in numpy.arange(0.0,1.1,0.1):
            predictions_Test = (predictions_test_content*alpha)+(predictions_test_context*(1-alpha))
            predictions_class_Test = predict_class(predictions_Test,threshold)
            report_performance_weighted_avg(file, time, "content and context", predictions_class_Test, type_class,alpha, threshold, predictions_Test[:,1])
```
